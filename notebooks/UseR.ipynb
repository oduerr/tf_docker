### How to call R code 

```python
import numpy as np
```

```python
x = np.linspace(0,2*3.14,100)
y = np.sin(x)
```

```python
%load_ext rpy2.ipython
```

```python
%Rpush x
%Rpush y
```

```python
%%R
#install.packages('tidyverse', repos="https://cloud.r-project.org")
library(ggplot2)
df = data.frame(x=x,y=y)
ggplot(df,aes(x=x,y=y)) + geom_point() 
```
